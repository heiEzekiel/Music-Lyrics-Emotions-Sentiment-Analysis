## Creating corpus for lyrics

```python
import nltk
from nltk.corpus.reader import PlaintextCorpusReader

# Create a list of strings
#my_list = ["This is the first document.", "This is the second document.", "This is the third document."]

# Create a temporary directory to store the corpus
corpus_dir = os.path.join(data_dir, "lyrics_corpus")
corpus_filename = 'my_corpus.txt'

# create text file for each row in lyrics_df
lyrics_list = lyrics_df["lyrics"].to_list() 
genre_list = lyrics_df["genre"].to_list()
topic_list = lyrics_df["topic"].to_list()
```

```python
import re
for i in range(len(lyrics_list)):
    corpus_filename = genre_list[i] + "_" + str(i) + ".txt"
    with open(os.path.join(corpus_dir, corpus_filename), 'w') as f:
        #remove special characters
        lyrics = re.sub('[^a-zA-Z0-9 \n\.]', '', lyrics_list[i])
        f.write(lyrics)
```

# Loading lyrics documents from corpus

```python
from preprocess2 import *

corpus = load_corpus(os.path.join(data_dir, "lyrics_corpus"))

docs = corpus2docs(corpus)

dictionary = gensim.corpora.Dictionary(docs)

vecs = docs2vecs(docs, dictionary)
print(len(docs))
```

```python
from k_means import *

num_tokens = len(dictionary.token2id)
clusters = k_means(vecs, num_tokens, 7)

```

```python
fids = corpus.fileids()

#The below prints the file ids in each cluster

cluster1 = clusters[0]
print("Cluster 1:", [fids[d] for d in cluster1])

cluster2 = clusters[1]
print("Cluster 2:", [fids[d] for d in cluster2])

cluster3 = clusters[2]
print("Cluster 3:", [fids[d] for d in cluster3])

cluster4 = clusters[3]
print("Cluster 4:", [fids[d] for d in cluster4])

cluster5 = clusters[4]
print("Cluster 5:", [fids[d] for d in cluster5])

cluster6 = clusters[5]
print("Cluster 6:", [fids[d] for d in cluster6])

cluster7 = clusters[6]
print("Cluster 7:", [fids[d] for d in cluster7])

```

```python
# Write your code here.

# Take all the file IDs in cluster1
cluster1_fids = [fids[d] for d in cluster1]

# Create an empty list clust1_words =[]
cluster1_all_words = []

# Add the words from all files to this list. Use corpus.words and extend method
for fid in cluster1_fids:
    cluster1_all_words.extend(corpus.words(fid))

    
#
# Remove the Stopwords from this list

stop_list = nltk.corpus.stopwords.words('english')

all_words1 = [w.lower() for w in cluster1_all_words]
all_words2 = [w for w in all_words1 if w not in stop_list and len(w)>4]

#Call freq distribution metod and display top 10 words or 20 words
fdist = nltk.FreqDist(all_words2)
print(fdist.most_common(20))
```
